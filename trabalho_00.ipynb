# Trabalho 0 - Transformações Lineares
## Funções utilizadas:

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_graph(name: str, vectors):
    plt.figure(figsize=(6, 6))

    for vec in vectors:
        plt.quiver(0, 0, vec["x"], vec["y"], angles='xy', scale_units='xy', scale=1, color=vec["color"], label=vec["name"])

    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.xticks(range(-5, 6))
    plt.yticks(range(-5, 6))

    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.grid(True)

    plt.legend()
    plt.xlabel("X", fontsize=12)
    plt.ylabel("Y", fontsize=12)
    plt.title(name, fontsize=14, fontweight='bold')

    plt.show()

def transform_vector(vector, matrix):
    return matrix @ vector

def create_shear_matrix(k):
    return np.array([[1, k], [0, 1]])
```

# Rotação horária de 90°

### Vetores originais:

$$
\vec{A} =
\begin{bmatrix}1 \\ 0\end{bmatrix}, \quad
\vec{B} =
\begin{bmatrix}0 \\ 1\end{bmatrix}
$$

---

### Novas posições após rotação horária de 90°:

$$
\vec{A}' =
\begin{bmatrix}0 \\ -1\end{bmatrix}, \quad
\vec{B}' =
\begin{bmatrix}1 \\ 0\end{bmatrix}
$$

---

### Matriz de transformação

Para obter a matriz, colocamos os vetores transformados como **colunas**:

$$
M =
\begin{bmatrix}
0 & -1 \\
1 & 0
\end{bmatrix}
$$

---

### Representação gráfica

- **Vetor A**: original $[1,0]$ → transformado $[0,-1]$  
- **Vetor B**: original $[0,1]$ → transformado $[1,0]$  

O gráfico mostra **como os vetores giram 90° no sentido horário**.


```python
initial_A = np.array([1, 0])
initial_B = np.array([0, 1])

matrix = np.array([[0, 1], [-1, 0]])

final_A = transform_vector(initial_A, matrix)
final_B = transform_vector(initial_B, matrix)

draw_graph("Rotação Horária de 90°",
    [
        #vetores unitários originais
        {"name": "A inicial", "x": initial_A[0], "y": initial_A[1], "color": "blue"},
        {"name": "B inicial", "x": initial_B[0], "y": initial_B[1], "color": "red"},

        #vetores rotacionados
        {"name": "A final", "x": final_A[0], "y": final_A[1], "color": "green"},
        {"name": "B final", "x": final_B[0], "y": final_B[1], "color": "orange"}
    ]
)
```

# Shear (Cisalhamento) no Eixo X

### Vetores originais:

$$
\vec{A} =
\begin{bmatrix}1 \\ 0\end{bmatrix}, \quad
\vec{B} =
\begin{bmatrix}0 \\ 1\end{bmatrix}
$$

---

### Aplicando o shear com $k = 2$ no eixo X:

As novas posições dos vetores unitários são:

$$
\vec{A}' =
\begin{bmatrix}1 \\ 0\end{bmatrix}, \quad
\vec{B}' =
\begin{bmatrix}2 \\ 1\end{bmatrix}
$$

---

### Matriz de transformação

A matriz é obtida colocando os vetores transformados como **colunas**:

$$
M =
\begin{bmatrix}
1 & 2 \\
0 & 1
\end{bmatrix}
$$

---

### Representação gráfica:

- **Vetor A**: original $[1,0]$ → transformado $[1,0]$  
- **Vetor B**: original $[0,1]$ → transformado $[2,1]$  

O gráfico mostra como o shear **desloca os vetores horizontalmente** proporcionalmente ao y deles.


```python
initial_A = np.array([1, 0])
initial_B = np.array([0, 1])

k = 2
matrix = create_shear_matrix(k)

final_A = transform_vector(initial_A, matrix)
final_B = transform_vector(initial_B, matrix)

draw_graph("Shear no Eixo X",
    [
        #vetores unitários originais
        {"name": "A inicial", "x": initial_A[0], "y": initial_A[1], "color": "blue"},
        {"name": "B inicial", "x": initial_B[0], "y": initial_B[1], "color": "red"},

        #vetores cerrilhados
        {"name": "A final", "x": final_A[0], "y": final_A[1], "color": "green"},
        {"name": "B final", "x": final_B[0], "y": final_B[1], "color": "orange"}
    ]
)
```
